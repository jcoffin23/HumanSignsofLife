**(a)** Listed below are the imports needed for this computer exercise.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from scipy.stats import multinomial
from sklearn.mixture import GaussianMixture as GMM 
```

**(b)** Here is the code to create your data set $(X_1,y_1)$, which are samples for class $\omega _1$.  In part (f) you are to create a second data set, $(X_1,y_1)$, with different parameters that correspond to class $\omega _2$.  If you wish, you may want to create both sets here.

One of the key steps in generating this data is creating the $n\times 3$ array `z`.  You may wish to print this array to see what it looks like, which may help you better understand how the mixture of Gaussians is created.  You may do this with the simple command `print(z)'.

```python
#
# Define the three Gaussian means, covariances, and mixing coefficients
#
mu_true = np.array([[1.25,1.25],[2.75,2.75],[2,6]])
sigma_true = np.array([[[0.1,0],[0,0.1]],[[0.2,0],[0,0.2]],[[0.3,0],[0,0.3]]])
lambda_true = np.array([.4, .4, .2]);
#
# Create an nx3 vector that has a one in the kth position with probabilities given
# in the array lambda_true.
#
n=500;
z = multinomial.rvs(1, lambda_true, size=n);
#
# Draw samples from the Gaussian mixture and
# create an array, y, containing the class label for each sample (in this case, 1)

X1 = np.array([np.random.multivariate_normal(mu_true[np.argmax(i)],
         sigma_true[np.argmax(i)]) for i in z])
y1=np.ones((n,), dtype=int) 
```

If you wish to plot the data samples that you generated, the following code will do this.

```python
import matplotlib.pyplot as plt
x1, x2 = X1.T
plt.plot( x1, x2, 'x', label='x(2)' )
plt.legend(loc='lower right')
plt.title('GMM, 500 sample points')
plt.show()
```

**(c)** The following is the statement used to split your data set into a *training set* and a *test set*.

```python
X1_train, X1_test,y1_train,y1_test=train_test_split(X1,y1,test_size=0.25) 
```

**(d)** The following commands will create a plot of the AIC for GMM using from one to 20 Gaussians.  You may wish to change this range of numbers.  You will also want to create a similar plot for the BIC.

```python
n_components = np.arange(1,10)
models = [GMM(n, covariance_type='full', random_state=0)
for n in n_components]
aics = [model.fit(X1).aic(X1) for model in models]
bics = [model.fit(X1).bic(X1) for model in models]
plt.plot(n_components, aics); 
plt.plot(n_components,bics);
plt.title('Aics and Bics for a number of models for FULL covariance')
plt.xlabel('Number of models')
plt.ylabel('Aic and Bic values')
```

# Discussion 

The AIC and BIC functions change drasticly if the covariance_type is changed. The defualt vavlue is 'full', which lets each model have its own full covariance matrix. If we change it to 'diag', then each model will have its own diagonal matrix. Doing this does not change the assumption that 3 models should be used. Above is the AIC and BIC using the 'Diag','spherical' and 'tied' covariance.

There is not much to say about using the diag or spherical covariances. The diag mode lets each model have its own diagional covariance matrix. Meaning that all models are assumed to have no cross correlation between variables and thus create distributions that are locked to spread along the x axis. The spherical mode uses a single covariance value. This will create gaussian distributions that are spherical. They produce results that allows the same conclusion as the 'full' case. They do not produce as sharp of peaks, so for this reason we would prefer the 'full' covariance mode.

The 'Tied' covarinace covariance type forces all models to share a single covariance matrix. This creates distributions that are exactly the same and only differ based on the mean.  This is non-optimal for this set of data. Looking at the plot below we can see that the number of models cannot be clearly taken from the AIC or BIC plots. There is no clearly well defined min value as there was in the 'Full' case.



For these reasons we would prefer the 'Full' covariance mode.

Here there is also an example of how to print out the GMM weights of the GMM.

```python
# Fit a Gaussian mixture with EM using three components
#
gmmDiag = GMM(n_components=3,covariance_type='diag')
gmmsph = GMM(n_components=3,covariance_type='spherical')
gmmFull = GMM(n_components=3,covariance_type='full')
gmmTied = GMM(n_components=3,covariance_type='tied')

gmmDiag.fit(X1_train) 
gmmsph.fit(X1_train)
gmmFull.fit(X1_train)
gmmTied.fit(X1_train)

mixCoe = [.2,.4,.4]


gds = np.sort(gmmDiag.weights_)
gss = np.sort(gmmsph.weights_)
gfs = np.sort(gmmFull.weights_)
gts = np.sort(gmmTied.weights_)

errorDiag = np.abs(mixCoe - gds)
errorSph = np.abs(mixCoe - gss)
errorFull = np.abs(mixCoe - gfs)
errorTied = np.abs(mixCoe - gts)

print("Mixxing coefficient errors")
print('Error for Diag Cov Mode: \n',errorDiag,'\n')

print('Error for Spherical Cov Mode: \n',errorSph,'\n')
print('Error for Tied Cov Mode: \n',errorTied,'\n')
print('Error for Full Cov Mode: \n',errorFull,'\n')




```

```python


print('Mean for Diag Cov: \n', gmmDiag.means_)
print('\n \n Mean for Spherical Cov: \n', gmmsph.means_)
print('\n \n Mean for Tied Cov: \n', gmmTied.means_)
print('\n \n Mean for Full Cov: \n', gmmFull.means_)
```

```python
print('Covariance matrix for Diag Cov: \n', gmmDiag.covariances_)
print('\n \n Covariance matrix for Spherical Cov: \n', gmmsph.covariances_)
print('\n \n Covariance matrix for Tied Cov: \n', gmmTied.covariances_)
print('\n \n Covariance matrix for Full Cov: \n', gmmFull.covariances_)
```

For the X_2 training, two gaussians are created centered at (2,6) and (2,2) where both are spherical. To Fix this, we can set the covariance mode to 'Full', which will allow the GMM to create a Distrubuiton that is still centered at (2,2), but is able to strech out in a 45 degree angle to align more with the data points.

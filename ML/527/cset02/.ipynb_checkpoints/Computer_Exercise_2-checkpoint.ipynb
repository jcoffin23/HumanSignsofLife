### Load the MNIST Dataset ###

```python
#Load the Dataset

from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1, cache=True)
mnist.target = mnist.target.astype(np.int8) # fetch_openml() returns targets as strings


```

```python
mnist.target
```

# Discussion Below
The Data of the data is 7000 by 784. This is because there are 7000 pictures in this data set. Each picture is a 28 by 28 image which has 784 pixels. These pixels form the 784 features in the feature vectors. The Target vector is vector of length 7000. Each value in the target vector is a number from 0 to 9 that identifies the character depicted in the image. 


### Put the dataset into a Data Array and a Target Array ###

```python
X, y = mnist["data"], mnist["target"]
print("MNIST Data Shape:", mnist.data.shape)
print("MNIST Target Shape:", mnist.target.shape)
```

### Display a Data Sample as an Image ### 

```python
number =  60000 # Put in some number here between 0 and 69,999
some_digit = X[number]
some_digit_image = some_digit.reshape(28, 28)
plt.imshow(some_digit_image, cmap=plt.cm.binary, interpolation="nearest")
plt.axis("off")
plt.title('Label = {}'.format(y[number]), fontsize = 15)
```

### Create a Training Set and a Test Set ### 

```python
X_train,X_test,y_train,y_test = X[:60000],X[60000:],y[:60000],y[60000:]
```

### Shuffle the Data ###

```python
shuffle_index = np.random.permutation(60000)
X_train,y_train = X_train[shuffle_index],y_train[shuffle_index]
```

### Create the Targets for an OVA Classifier ###

```python
n8 = 8    # Set n equal to some number here between 0 and 9
y_train_n8 = (y_train == n8) # True for all n's, False for all other digits
y_test_n8  = (y_test == n8)
```

### OvA SGD Classifier ###

Use the SGDClassifer class with a *'perceptron'* loss to train an OvA classifer for your digit `n`.

```python
from sklearn.linear_model import SGDClassifier
clf8 = SGDClassifier(loss='perceptron',random_state=72,warm_start=True,alpha = 2)
clf8.fit(X_train, y_train_n8) 
```

### Evaluating the Classifier Performance ###

```python
coefs = clf8.coef_
cnt = 0
print(coefs.shape)
coeStd = np.std(coefs)
coeMean = np.mean(coefs)
for i in range(0,coefs.shape[1]):
    if abs(coefs[0,i]) > .1*coeStd: #Number of Coefs above 1/10 STD from zero
        cnt +=1
        
print(cnt)

prediction = clf8.predict(X_test[1:100])

correct = sum(prediction == y_test_n8[1:100])
print(correct)
```

### Effect of Training Sample Size ###

```python
regr = LogisticRegression(solver = 'lbfgs')
fig, axes = plt.subplots(nrows = 1, ncols = 3, figsize = (24,8));
plt.tight_layout()

for plotIndex, sample_size in enumerate([100, 1000, 60000]):
    X_train_ss = X_train[:sample_size].reshape(sample_size, 784)
    y_train_ss = y_train[:sample_size]
    regr.fit(X_train_ss, y_train_ss)
    predicted = regr.predict(X_test)
    cm = metrics.confusion_matrix(y_test, predicted)
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    sns.heatmap(cm_normalized, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r', ax = axes[plotIndex], cbar = False);
    accuracyString = '{:g} Training Samples Score: {:.3f}'.format(sample_size, regr.score(X_test, y_test)) 
    axes[plotIndex].set_title(accuracyString, size = 25);

axes[0].set_ylabel('Actual label', fontsize = 30);
axes[1].set_xlabel('Predicted label', fontsize = 30);
```

The code above clearly gives three results for the three values of N. The trend is predictiable. The larger the training set the higher the accuracy. There are some intresting points to consider though.  It appears that adding more samples to the training set, N = 100 to N =1000 does not lower the misclassifications of all classes equally. 
What I mean is that for class 5 the predicted label of 3 is higher when N = 1000. At this same jump in N the missclassification for class 2 fell across the predicted labels. This means that any increase in N does NOT mean that in general any misclassification will go down. However look when N = 60,000 and that same block of class 5 missclassified as 3 drops by more than half but not to zero.  This tells us that in general over a large enough increase of N the value of any one missclassification will go down, but it could increase in a small increase of N.

Another thing to note is that as N increases the decrease in error is diminishing. An increase of just 900 samples decreases the error by about 12% while an additional increase of 50,000 drops error by just 8%. This diminishing return tells us that it will be imposible to get a perfect classifier.
